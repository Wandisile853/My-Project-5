```python
#scatter plot
x= [10,20,30,40,50,60,70,80,90]
a= [1,2,3,4,5,4,3,2,9]
b= [5,6,7,8,3,2,1,4,9] 

plt.figure(figsize=(10,10)) 
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=500,marker="3")
plt.legend(['a','b']) 
plt.title('x vs a&b') 
plt.xlabel('x') 
plt.ylabel('a&b') 
plt.show() 
```

```python
#Histogram to see the relationship between age and wear specs 
plt.hist(wandi['Age'],color='red',bins=50) 
plt.show()
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three]) 
plt.violinplot(data,showmeans=True,showmedians=True)
plt.grid(True) 
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#pie chart, here we see which categories have how much percentage 
data=('Age','IQ','Sleeping hours','Number of friends') 
quantity=(30,150,8,5) 
plt.pie(quantity,labels=data,autopct='%0.1f%%',shadow=True,explode=(0,0,0.1,1))  
plt.show()
```

```python
#Donut plot which is just the same as a pie chart, only differance is the shape
data=('Age','IQ','Sleeping hours','Number of friends') 
quantity=(30,150,8,5) 

pie1=plt.pie(quantity,labels=data,autopct='%0.1f%%',radius=2)  
pie2=plt.pie([5],colors='w',radius=1)

plt.show() 
```

```python
#Area plot, similar to histogram, to show the numerical distribution of the data

x=[10,20,30,40,50,60,70,80,90]
y=[1,1,2,5,6,4,7,4,6]
plt.stackplot(x,y) 
plt.show() 
```

```python
#now we can plot a littte complicated plot, a joint plot, which is basically a scatter plot and a distribution plot at the same time
sns.jointplot(x="Age",y="IQ",data=wandi, kind="reg")
```
